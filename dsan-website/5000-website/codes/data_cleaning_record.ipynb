The data was then transposed so that the columns become rows and rows become columns.

```python
df = df.iloc[4:]
df = df.iloc[1:]
df.head()
```

The unnecessary information has been removed and the missing values are checked now

```python
df.isnull().sum()
```

In the dataset, there was a problem of the dataframe having row indexes and column indexes and hence, I worked on it drop the index and create a yearly column from scratch.

```python
# creating a new index rather than treating it
new_index = range(1979, 2022)

# Reset the index with the new range
df = df.reset_index(drop=True)
df.index = new_index
df.index.name = "year"

df.shape
```

All the columns have been renamed

```python
cols= ['adj_NNI_g', 'adj_NNI_c', 'adj_savings_fix_cap_GNI', 'adj_savings_edu_GNI', 'adj_NNS_GNI',
                    'curr_acc_bal', 'ex_imp_gdp', 'ex_imp_growth', 'ex_debt_shocks', 'final_consump_exp_gdp',
                    'final_consump_exp_growth', 'fdi_net_inflows', 'fdi_net_outflows', 'gdp_growth', 'gdp_per_capita_growth',
                    'gener_govt_fin_consump_exp', 'gni_growth', 'goods_exp_bop','goods_imp_bop', 'imports_goods_services_gdp', 'imports_goods_services_growth','services_gdp', 'services_growth', 
                       'short_term_debt_tot_reserves','trade_gdp', 'cpi', 'inflation','real_interest_rate','lending_interest_rate','life_exp_birth','expense_gdp','interest_payments', 'military expenditure']

df.columns = cols
df.head(5)
```

Used mean value imputation for missing value imputation. As discussed, for modelling missing values have to be removed and hence, this is the reason why they're being removed. But in general, for economic data, to prevent bias, the missing value imputation is not done at all as it introduces bias to the dataset.

```python
value_to_replace = ".."
for col in df.columns:
    df[col] = df[col].replace(value_to_replace, np.nan)

missing= df.isnull().sum()
missing
missing_vals = pd.DataFrame({'Missing Values': missing.index, 'Missing Values Sum': missing.values})
missing_vals = missing_vals.sort_values(by='Missing Values Sum', ascending=False)

print(missing_vals.head(15))
```

```python
for cols in df.columns:
    mean = df[cols].mean()
    df[cols] = df[cols].fillna(mean)
    
print(df.isnull().sum())
```

Market-based economies now days are centred on steady growth. A collection of interconnected production and consuming activities that decide how scarce resources are distributed is called an economy. These limited resources are distributed within a market system in market-based economies, where supply and demand dictate how resources are allocated. 

The BRICS countries have seen a great deal of economic growth over the last few decades. China, in particular, has rapidly industrialised and grown to be the second-largest economy in the world. India has also played a significant role in this growth. These countries have diverse economies with interconnected production and consumption activities. As a group, they have significant influence on the global stage. Their trade policies, economic activities, and strategic decisions have an impact not only on their individual economies but also on the global economy as a whole. For this reason, GDP growth rate is essential in gauging the overall performance of the economy.

According to various sources, the ideal growth rate for a developed economy is around 2-3% but again, for developing country, no literature has shown be a well-performing percentage for good growth rate or moderate growth rate. However, GDP though can be a good indicator for economic performance, focusing on other parameters of education, health, gender and sustained-demand have contribute for the better performance as an economy. Yet, keeping in mind the metrics seen for developed economies, the labels have been created accordingly. With 0-2% as Low Growth, 2-5% as Moderate Growth, 5-7% as 'High Growth' looking into the average growth rates of developing economies that are performing well and 7 and above is shown as 'Very High Growth'. This is user-defined for the purposes of predictive modelling. 

```python
bins = [0, 2, 5, 7, float('inf')]
labels = ['Low Growth', 'Moderate Growth', 'High Growth', 'Very High Growth']
df['labels'] = pd.cut(df['gdp_growth'], bins=bins, labels=labels)
df.head(5)
```
